## Analysing the Outliers

```python
Q1=df['Price'].quantile(0.25)
Q3=df['Price'].quantile(0.75)
IQR=Q3-Q1
lowoutlier=Q1-1.5*IQR
highoutlier=Q3+1.5*IQR
totaloutlier=((df['Price']<lowoutlier)|(df['Price']>highoutlier)).sum()
print('Total no of Outliers: ',totaloutlier)
print('Low Outlier: ',lowoutlier)
print('High Outlier: ',highoutlier)


# Printing the Total No of Outliers
# Printing the Low Outlier
# printing the High Outlier
```

## Importing the Libraries Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Plotting Histogram for the field 'Rating'

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Rating'],bins=50)

# This Histogram is plotted for the Count of Ratings from 0 to 5 star

# There are most number of books present in the rating range of 4 to 5
```

## Plotting Box Plot with respective 'Main Genre' and 'Rating'

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Rating', y='Main Genre', data=df, orient='h')
plt.title('Distribution of Ratings by Genre')
plt.xlabel('Rating')
plt.ylabel('Genre')
plt.show()

# This Box Plot is plotted to Find the Distribution of Genres among all the Star ratings
# We can observe that the 'LAW' Genre has books among all the ratings
```

## Plotting a Bar Graph for the Field 'Main Genre'

```python
plt.figure(figsize=(20, 6))
sns.countplot(x='Main Genre',data=df)
plt.xticks(rotation=80)
plt.show()

# Number of books present in each Main Genre
```

## Plotting the Bar Graph for the Field 'Type'

```python
plt.figure(figsize=(20, 6))
sns.countplot(x='Type',data=df)
plt.xticks(rotation=80)
plt.show()

# This barplot is plotted to find the count of the type of the books
# We can find out that The most number of books are in the type 'paperback' and 'Kindle'
```

## Plotting a BarPlot for Main Genre & Price

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Price', y='Main Genre', data=df, estimator=np.mean)
plt.title('Average Price by Genre')
plt.xlabel('Average Price')
plt.ylabel('Genre')
plt.show()

# Here the Barplot is plotted between 'Main Genre' and 'Price'
# Here we have taken estimation on the bases of Mean of prices
# This Bar Plot tells the average price of each 'Main Genre'
# From the BarPlot we can find that The books belong to 'Medicine & Health Sciences' are relatively expensive than the other Genres.
```

## Plotting a Bar Plot for 'Rating' & 'Main Genre'

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Rating', y='Main Genre', data=df, estimator=np.mean)
plt.title('Average Rating by Genre')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.show()

# This Bar Plot is plotted to find Averag Rating by Genre
# From this Bar Plot we can find out that Law Books have Less Average Rating among all the other books
```

## Ranking the Authors by 'Rating' and 'No.of People rated'

```python
highly_rated_books = df[df['Rating']>=4.5]

ratings_threshold = highly_rated_books['No. of People rated'].quantile(0.25)

hidden_gems = highly_rated_books[highly_rated_books['No. of People rated']<=ratings_threshold]

hidden_gems_sorted = hidden_gems.sort_values(by=['Rating','Price'],ascending=[False,True])

hidden_gems_sorted.head(10)
```

## Printing Top 10 Best rated Authors,Most Expensive Authors,Top Rated Authors

```python
min_ratings_threshold = 1000
author_stats = df.groupby('Author').agg({
    'Rating':'mean',
    'Price':'mean',
    'No. of People rated':['sum','mean']
}).reset_index()
author_stats.columns = ['Author','Avg Rating','Avg Price','Total Ratings','Avg Ratings per Book']
best_rated_authors = author_stats[author_stats['Avg Ratings per Book'] > min_ratings_threshold].sort_values(by='Avg Rating', ascending=False).head(10)
most_expensive_authors = author_stats.sort_values(by='Avg Price', ascending=False).head(10)
most_rated_authors = author_stats.sort_values(by='Total Ratings', ascending=False).head(10)

best_rated_authors,most_expensive_authors,most_rated_authors
```

## Plotting Bar plots for Best Rated Authors,Most Expensive Authors and Top rated Authors

```python
plt.figure(figsize=(18,14),dpi=200)

plt.subplot(3,1,1)
sns.barplot(x='Avg Rating', y='Author', data=best_rated_authors)
plt.title('Top 10 best Rated Authors')
plt.xlabel('Average Rating')
plt.ylabel('Author')

plt.subplot(3,1,2)
sns.barplot(x='Avg Price', y='Author', data=most_expensive_authors)
plt.title('Top 10 most Expnsive Authors')
plt.xlabel('Average Price (₹)')
plt.ylabel('Author')

plt.subplot(3,1,3)
sns.barplot(x='Total Ratings',y='Author',data=most_rated_authors)
plt.title('Top 10 Most Rated Authors')
plt.xlabel('Total Ratings')
plt.ylabel('Author')

plt.tight_layout()
plt.show()

# 3 Bar Plots are plotted
# Fist Bar Plot is for Top 10 Best rated Authors, 'Joy Alukkas' is the Best Rated Author
# Second Barplot is for Most Expensive Authors, 'Sigeru Omantu' is the most Expensive Author
# Third Barplot is for Most Rated Authors, 'J K.Rowling' is the Most Rated Author
```

## Plotting a Scatter Plot between 'Price' and 'Rating of Books'

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='Price',y='Rating',data=df,alpha=0.5)
plt.title('Price vs. Rating of Books')
plt.xlabel('Price (₹)')
plt.ylabel('Rating')
plt.xscale('log')
plt.grid(True,which="both",ls="--")
plt.show()

# From this Scatterplot we can say the books which are having rating from 4 to 5 are the most expensive
```

## DashBoard 1

```python
# Assuming df is your dataframe containing the data

# Create a figure and set of subplots
fig, axs = plt.subplots(2, 2, figsize=(24, 16))

# Histogram
sns.histplot(df['Rating'], bins=50, ax=axs[0, 0], color='skyblue')
axs[0, 0].set_title('Count of Ratings from 0 to 5 star', fontsize=16)
axs[0, 0].set_xlabel('Rating', fontsize=12)
axs[0, 0].set_ylabel('Count', fontsize=12)

# Box plot
sns.boxplot(x='Rating', y='Main Genre', data=df, orient='h', ax=axs[0, 1], palette='pastel')
axs[0, 1].set_title('Distribution of Ratings by Genre', fontsize=16)
axs[0, 1].set_xlabel('Rating', fontsize=12)
axs[0, 1].set_ylabel('Genre', fontsize=12)

# Count plot for Main Genre
sns.countplot(x='Main Genre', data=df, ax=axs[1, 0], palette='muted')
axs[1, 0].set_title('Number of books in each Main Genre', fontsize=16)
axs[1, 0].set_xlabel('Main Genre', fontsize=12)
axs[1, 0].set_ylabel('Count', fontsize=12)
axs[1, 0].tick_params(axis='x', rotation=45)

# Count plot for Type
sns.countplot(x='Type', data=df, ax=axs[1, 1], palette='bright')
axs[1, 1].set_title('Count of the type of books', fontsize=16)
axs[1, 1].set_xlabel('Type', fontsize=12)
axs[1, 1].set_ylabel('Count', fontsize=12)
axs[1, 1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Add a title for the entire dashboard
plt.suptitle('Dashboard 1', fontsize=24, y=1.02)

# Add some style
sns.set_style('whitegrid')

# Show the plots
plt.show()
```

## Dashboard 2

```python
# Assuming df is your dataframe containing the data

# Create a figure and set of subplots
fig, axs = plt.subplots(1, 2, figsize=(18, 8))

# Barplot of Average Price by Genre
sns.barplot(x='Price', y='Main Genre', data=df, estimator=np.mean, ax=axs[0], palette='Blues_r')
axs[0].set_title('Average Price by Genre', fontsize=16)
axs[0].set_xlabel('Average Price', fontsize=12)
axs[0].set_ylabel('Genre', fontsize=12)

# Barplot of Average Rating by Genre
sns.barplot(x='Rating', y='Main Genre', data=df, estimator=np.mean, ax=axs[1], palette='Greens_r')
axs[1].set_title('Average Rating by Genre', fontsize=16)
axs[1].set_xlabel('Average Rating', fontsize=12)
axs[1].set_ylabel('Genre', fontsize=12)

# Add annotations to the bars
for ax in axs:
    for p in ax.patches:
        ax.annotate(f"{p.get_width():.2f}", ((p.get_width() * 1.005), p.get_y() + p.get_height() / 2), ha='left', va='center', color='black', fontsize=10)

# Adjust layout
plt.tight_layout()

# Add a title for the entire dashboard
fig.suptitle('Dashboard 2 - Average Price and Rating by Genre', fontsize=24)

# Show the plots
plt.show()
```
